````python
# Revisando lo de 30 (Meteorite)255 (Core)
columnas_deseadas = ['capture_rate', 'classfication', 'name', 'type1', 'type2']
valor_captura_especifico = "30 (Meteorite)255 (Core)"

# Filtra las filas que cumplen con la condición en la columna 'capture_rate'
data_subset = data[data['capture_rate'] == valor_captura_especifico][columnas_deseadas]
data_subset
````

**Formas**

Cuando su coraza se quiebra, el núcleo de Minior queda al descubierto, revelando así el color que esconde en su interior:

Minior.png

Así que Core significa sólo que se sabe su color porque su coraza se quebró.

Entonces se deben sumar el 30 de Meteorite y los 255 de Core.

Minior capture_rate = 285


````python
valor_original = "30 (Meteorite)255 (Core)"
nuevo_valor = "285"

# Actualiza el valor en la columna 'capture_rate'
data.loc[data['capture_rate'] == valor_original, 'capture_rate'] = nuevo_valor
````

## Dejar las columnas que se usarán

````python
data.columns
````

````python
columnas_deseadas = ['against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water', 'attack',
       'base_egg_steps', 'base_happiness', 'base_total', 'capture_rate',
       'defense', 'experience_growth', 'height_m', 'hp', 'sp_attack',
       'sp_defense', 'speed', 'type1', 'weight_kg', 'is_legendary']
data = data.loc[:, columnas_deseadas]
data.head()
````

# Preprocesamiento

is_legendary es la única columna binaria (0 y 1)

````python
# Valores únicos para cada columna
unique_values = data.nunique()
# Encuentra las columnas con 2 valores únicos
binary_columns = unique_values[unique_values == 2].index

print(binary_columns)
````

La variable objetivo en type_1, es categórica y se convertirá a números con LabelEncoder

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Aplicar LabelEncoder a la variable objetivo
data['type1'] = le.fit_transform(data['type1'].values)
````

````python
data.head(5)
````

# Separación de datos

````python
X = data[['against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water', 'attack',
       'base_egg_steps', 'base_happiness', 'base_total', 'capture_rate',
       'defense', 'experience_growth', 'height_m', 'hp', 'sp_attack',
       'sp_defense', 'speed', 'weight_kg', 'is_legendary']]
y = data['type1'].values
````

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Establece los valores de las columnas a estandarizar
to_scale = list(X.columns)
to_scale.remove('is_legendary')

# Estandariza las columnas
X[to_scale] = scaler.fit_transform(X[to_scale])
````

````python
X.describe()
````

````python
from sklearn.model_selection import train_test_split
# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=123)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

# Modelo, Entrenamiento y Predicciones

## XGBoost

````python
from sklearn.model_selection import KFold, GridSearchCV
import xgboost as xgb

# Dividir los datos en k-folds
kf = KFold(n_splits=5, shuffle=True, random_state=123)

# Definir el modelo XGBoost
model_xgb = xgb.XGBClassifier()

# Definir el espacio de búsqueda para los hiperparámetros
param_grid = {
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7]
}

# Inicializar el GridSearchCV
grid_search = GridSearchCV(estimator=model_xgb, param_grid=param_grid, scoring='accuracy', cv=kf)

# Realizar la búsqueda de hiperparámetros utilizando los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Imprimir los mejores hiperparámetros encontrados
print("Mejores hiperparámetros:", grid_search.best_params_)

# Obtener el mejor modelo entrenado
best_model = grid_search.best_estimator_

# Evaluar el modelo en los datos de prueba
pred_xgb= best_model.predict(X_test)
````

### Métricas

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy_xgb = accuracy_score(y_test, pred_xgb)
precision_xgb = precision_score(y_test, pred_xgb, average='macro', zero_division='warn')
recall_xgb = recall_score(y_test, pred_xgb, average='macro')
f1_xgb = f1_score(y_test, pred_xgb, average='macro')
````

````python
print('El accuracy con XGBoost fue: ', accuracy_xgb)
print('La precisión con XGBoost fue: ', precision_xgb)
print('El recall con XGBoost fue: ', recall_xgb)
print('f1 Score con XGBoost fue: ', f1_xgb)
````

````python
from sklearn.metrics import classification_report

report_xgb = classification_report(y_test, pred_xgb)
print(report_xgb)
````

## Random Forest

````python
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Dividir los datos en k-folds
kf = KFold(n_splits=5, shuffle=True, random_state=123)

# Definir el modelo Random Forest
model_rf = RandomForestClassifier(n_estimators=100, max_depth=10)

# Definir el espacio de búsqueda para los hiperparámetros
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Inicializar el GridSearchCV
grid_search_rf = GridSearchCV(estimator=model_rf, param_grid=param_grid, scoring='accuracy', cv=kf)

# Realizar la búsqueda de hiperparámetros utilizando los datos de entrenamiento
grid_search_rf.fit(X_train, y_train)

# Imprimir los mejores hiperparámetros encontrados
print("Mejores hiperparámetros:", grid_search_rf.best_params_)

# Obtener el mejor modelo entrenado
best_model_rf = grid_search_rf.best_estimator_

# Evaluar el modelo en los datos de prueba
pred_rf = best_model_rf.predict(X_test)

# Puedes seguir con el resto de tu código utilizando el mejor modelo, por ejemplo:
# métricas, matriz de confusión, etc.
````

## Logistic Regression

````python
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Dividir los datos en k-folds
kf = KFold(n_splits=3, shuffle=True, random_state=123)

# Definir el modelo de Regresión Logística
model_lr = LogisticRegression(max_iter=1000)

# Definir el espacio de búsqueda para los hiperparámetros
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Parámetro de regularización
    'penalty': ['l1', 'l2','elasticnet']  # Tipo de penalización
}

# Inicializar el GridSearchCV
grid_search_lr = GridSearchCV(estimator=model_lr, param_grid=param_grid, scoring='accuracy', cv=kf)

# Realizar la búsqueda de hiperparámetros utilizando los datos de entrenamiento escalados
grid_search_lr.fit(X_train, y_train)

# Imprimir los mejores hiperparámetros encontrados
print("Mejores hiperparámetros:", grid_search_lr.best_params_)

# Obtener el mejor modelo entrenado
best_model_lr = grid_search_lr.best_estimator_

# Evaluar el modelo en los datos de prueba escalados
pred_lr = best_model_lr.predict(X_test)
````

### Métricas

````python
accuracy_lr = accuracy_score(y_test, pred_lr)
recall_lr = recall_score(y_test, pred_lr, average='macro')
precision_lr = precision_score(y_test, pred_lr, average='macro')
f1_lr = f1_score(y_test, pred_lr, average='macro')

print('El accuracy con LogisticRegression fue: ', accuracy_lr)
print('La precisión con LogisticRegression fue: ', precision_lr)
print('El recall con LogisticRegression: ', recall_lr)
print('f1 Score con LogisticRegression fue: ', f1_lr)
````

````python
from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(y_test, pred_lr)
fig, ax = plt.subplots(figsize=(10,10))
ax.matshow(confmat, cmap=plt.cm.Blues, alpha=0.3)
for i in range(confmat.shape[0]):
  for j in range(confmat.shape[1]):
    ax.text(x=j, y=i, s=confmat[i,j],
            va='center', ha='center')
plt.xlabel('Clase predicha')
plt.ylabel('Clase real')
plt.show()
````

````python
# Genera el reporte de clasificación
report_lr = classification_report(y_test, pred_lr)
print(report_lr)
````
