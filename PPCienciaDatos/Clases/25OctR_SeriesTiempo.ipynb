```r
## funcion para saber si la variable esta autocorrelacionada
## si las lineas verticales estan dentro de las lineas punteadas entonces si estan correlacionadas
acf(economics$pop,50)
```

```r
acf(economics$psavert,50)
```

```r
p <- ggplot(data=economics, aes(x=date,y=psavert))+
  geom_line(linewidth=1)
## grafica de todos los datos
p

#grafica con menos datos
min <- as.Date('2002-1-1')
max <- NA
p + scale_x_date(limits=c(min,max))
p + stat_smooth(color='#FC4E07',method='loess')
```

```r
library('TTR')
p <- ggplot(data=economics, aes(x=date,y=psavert))+
  geom_line(linewidth=1)
## grafica de todos los datos
p

#grafica con menos datos
min <- as.Date('2002-1-1')
max <- NA
p + scale_x_date(limits=c(min,max))
p + stat_smooth(color='#FC4E07',method='loess')
psr <- SMA(economics,n=8)
p + geom_line(aes(x=date,y=psr),color='#00AFBB')
```
