```python
import matplotlib.pyplot as plt
from math import sqrt
def heron(N, n=1, e=1e-5):
  """
  heron: función que calcula raíces con el método de Heron de Alejandría
  Parametros:
  N: numero (positivo)
    Número del cual deseamos calcular la raíz
  n: número (positivo)
     Valor inicial para la posible raíz
  e: flotante (positivo)
    Error máximo aceptable
  """
  if N<=0: return None
  err, i=1.0, 0
  while err > e:
    print(i,'  => ', n, ' error : ', err)
    i += 1
    plt.scatter(i, n, c='green', s=10)
    plt.scatter(i, err, c='red', s=10)
    nn = 0.5 * (N(n+n))
    err = abs(n-nn)/n
    n = nn
    plt.xlabel('Iteración')
    plt.ylabel('Raíz (verde) / Error (rojo)')
    return n
```

```python
N,n = 100,5
heron(N,n)
print('sqrt => ',sqrt(N))
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.style.use('ggplot')
```

```python
def Biseccion(f,a0,b0,emax=1e-5,iter_max=50,graf=False):
  a,b = a0, b0
  error, i = 10,0
  C, Fc = [],[]
  while error > emax and i < iter_max:
    c = (a+b)/2
    fa = f(a)
    fc = f(c)
    C.append(c)
    Fc.append(fc)
    if fa * fc < 0: b = c
    else: a = c
    error = abs(fc)
    i += 1
  if graf: grafica(f,C,Fc)
  return i,c
#f function
#C centroides
#fc conjunto de evaluaciones

def grafica(f,c,fc):
  cc = c.copy()
  cc.sort()
  x = np.linspace(cc[0],cc[-1],100)
  plt.plot(x,f(x),c='green')
  plt.scatter(c,fc,c='red',s=10)
  plt.scatter(c[-1],fc[-1],c='blue',marker='*',s=40)
  plt.show()

# el menor de todos los centroides es el primero
```

```python
def f1(x):
  return x**3 -2*x**2 -1

x = np.linspace(-1,5,100)
plt.plot(x,f1(x))
plt.show()
```

```python
iter,raíz = Biseccion(f1,1,2.5,graf=True)
print(iter,raíz)
```

esto es mejor que la de heron porque ya tenemos la grafica

```python
def f2(x):
  return np.sin(2*x) * (x**3 -2*x**2 -1)

x = np.linspace(-4,4,100)
plt.plot(x,f2(x))
plt.show()
```

```python
iter,raíz = Biseccion(f2,-4,-3,graf=True)
print(iter,raíz)
```

Método Newton-Raphson

```python
def Df(f,x,h=1e-3):
  return(f(x+h)-f(x))/h

#def
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,4*np.pi,100)
f = np.sin
fdy = np.cos

plt.plot(x,f(x),label='FO')
plt.plot(x,fdy(x),label='DS', marker='o')
plt.plot(x,Df(f,x),label='DN')
plt.legend()
plt.show()
```

```python
def newton_raphson(f,x0,emax=1e-5,iter_max=50,graf=True):
  error, i = 10,0
  C,Fc = [],[]
  while error > emax and i < iter_max:
    xi = x0 - f(x0)/Df(f,x0)
    C.append(x0)
    Fc.append(f(x0))
    error = abs(f(xi)-f(x0))
    x0 = xi
    i += 1
  if graf: grafica(f,C,Fc)
  return i,x0
```

```python
iter,raíz = newton_raphson(f1,1.6)
print(iter,raíz)
```

```python
iter,raíz = newton_raphson(f2,-3.7)
print(iter,raíz)
```
