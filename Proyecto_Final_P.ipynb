## Redimensionar

Las imagenes se encuentran en diferentes tamaños por lo que vamos a cambiar las dimensiones de todas las imagenes seleccionadas a las mismas para evitar tener problemas de dimencion en el arreglo en los pasos posteriores

### Redimensionando las imágenes de perros:

````python
# NO CORRER ESTA PARTE SI YA SE REDIMENSIONARON LAS IMÁGENES EN UNA NUEVA CARPETA!!!!!
# Ruta de la carpeta original con las imágenes de perros
input_folder = "/content/drive/My Drive/Patrones/Dog"

# Ruta de la nueva carpeta para guardar las imágenes redimensionadas
output_folder = "/content/drive/My Drive/Patrones/Perros"
os.makedirs(output_folder, exist_ok=True)

# Lista de nombres de archivos en la carpeta original
all_images = os.listdir(input_folder)

# Selecciona aleatoriamente n imágenes
n = 3000
selected_images = random.sample(all_images, min(n, len(all_images)))

# Redimensiona y guarda las imágenes en la nueva carpeta
for img_name in selected_images:
    # Verifica si el archivo tiene una extensión de imagen conocida
    if img_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        input_path = os.path.join(input_folder, img_name)
        output_path = os.path.join(output_folder, img_name)

        # Lee la imagen original
        original_image = Image.open(input_path)

        # Verifica el modo de color y convierte a 'RGB' si es necesario
        if original_image.mode != 'RGB':
            original_image = original_image.convert('RGB')

        # Redimensiona la imagen a x pixeles
        resized_image = original_image.resize((140, 140))

        # Guarda la imagen redimensionada en la nueva carpeta
        resized_image.save(output_path)

print("Proceso completado")
````

## Arquitectura de la Red Neuronal

````python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Activation, Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import preprocess_input
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.optimizers import Adam
import time
````

````python
def arquitectura_modelo(activation, input_shape, dropout):
  model_1 = Sequential()

# Añadimos la primera capa
  model_1.add(layers.Conv2D(50,(3,3), activation = activation, input_shape = input_shape))
  model_1.add(layers.MaxPooling2D(pool_size = (2,2)))
  model_1.add(Dropout(dropout))
# Añadimos la segunda capa
  model_1.add(layers.Conv2D(100,(3,3), activation = activation))
  model_1.add(layers.MaxPooling2D(pool_size = (2,2)))
  model_1.add(Dropout(dropout))

# Capa Flatten para aplanar las características antes de la capa densa
  model_1.add(Flatten())

# Capa densa con activación 'relu'
  model_1.add(Dense(100, activation=activation))

# Capa de salida con activación 'sigmoid' para clasificación binaria
  model_1.add(Dense(1, activation='sigmoid'))

  return model_1
````

````python
def arquitectura_modelo2(activation, input_shape):
  model_2 = Sequential()

# Añadimos la primera capa
  model_2.add(layers.Conv2D(32,(3,3), activation = activation))
  model_2.add(layers.MaxPooling2D(pool_size = (2,2)))

# Capa Flatten para aplanar las características antes de la capa densa
  model_2.add(Flatten())

# Capa densa con activación 'relu'
  model_2.add(Dense(64, activation=activation))

# Capa de salida con activación 'sigmoid' para clasificación binaria
  model_2.add(Dense(1, activation='sigmoid'))

  return model_2
````

````python
def arquitectura_modelo3(activation, input_shape):
  model_3 = Sequential()

# Añadimos la primera capa
  model_3.add(layers.Conv2D(32,(3,3), activation = activation))
  model_3.add(layers.MaxPooling2D(pool_size = (2,2)))
# Añadimos la segunda capa
  model_3.add(layers.Conv2D(64,(3,3), activation = activation))
  model_3.add(layers.MaxPooling2D(pool_size = (2,2)))

# Capa Flatten para aplanar las características antes de la capa densa
  model_3.add(Flatten())

# Capa densa con activación 'relu'
  model_3.add(Dense(32, activation=activation))

# Capa de salida con activación 'sigmoid' para clasificación binaria
  model_3.add(Dense(1, activation='sigmoid'))

  return model_3
````

````python
def arquitectura_modelo4(activation, input_shape):
  model_4 = Sequential()

# Añadimos la primera capa
  model_4.add(layers.Conv2D(32,(3,3), activation = activation, input_shape = input_shape))
  model_4.add(layers.MaxPooling2D(pool_size = (2,2)))
# Añadimos la segunda capa
  model_4.add(layers.Conv2D(64,(3,3), activation = activation))
  model_4.add(layers.MaxPooling2D(pool_size = (2,2)))
# Añadimos la tercera capa
  model_4.add(layers.Conv2D(128,(3,3), activation = activation))
  model_4.add(layers.MaxPooling2D(pool_size = (2,2)))
# Añadimos la cuarta capa
  model_4.add(layers.Conv2D(256,(3,3), activation = activation))
  model_4.add(layers.MaxPooling2D(pool_size = (2,2)))
# Capa Flatten para aplanar las características antes de la capa densa
  model_4.add(Flatten())
# Capa de salida con activación 'sigmoid' para clasificación binaria
  model_4.add(Dense(1, activation='sigmoid'))

  return model_4
````

````python
def arquitectura_modelo5(activation, input_shape, dropout):
  model_5 = Sequential()

# Añadimos la primera capa
  model_5.add(layers.Conv2D(32,(3,3), activation = activation, input_shape = input_shape))
  model_5.add(layers.MaxPooling2D(pool_size = (2,2)))
  model_5.add(Dropout(dropout))
# Añadimos la segunda capa
  model_5.add(layers.Conv2D(64,(3,3), activation = activation))
  model_5.add(layers.MaxPooling2D(pool_size = (2,2)))
  model_5.add(Dropout(dropout))
# Añadimos la tercera capa
  model_5.add(layers.Conv2D(128,(3,3), activation = activation))
  model_5.add(layers.MaxPooling2D(pool_size = (2,2)))
  model_5.add(Dropout(dropout))
# Añadimos la cuarta capa
  model_5.add(layers.Conv2D(256,(3,3), activation = activation))
  model_5.add(layers.MaxPooling2D(pool_size = (2,2)))
# Capa Flatten para aplanar las características antes de la capa densa
  model_5.add(Flatten())
# Capa de salida con activación 'sigmoid' para clasificación binaria
  model_5.add(Dense(1, activation='sigmoid'))

  return model_5
````

````python
# Función para mostrar las gráficas de pérdida y accuracy por época

def plot_loss_and_accuracy(history):
    # Crear un conjunto de subgráficas con 2 filas y 1 columna
    plt.figure(figsize=(12, 4))

    # Primera gráfica: Pérdida
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Gráfica de Pérdida')
    plt.ylabel('Costo')
    plt.xlabel('Época')
    plt.legend(['Entrenamiento', 'Validación'], loc='upper right')

    # Segunda gráfica: Precisión
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Gráfica de Precisión')
    plt.ylabel('Precisión')
    plt.xlabel('Época')
    plt.legend(['Entrenamiento', 'Validación'], loc='lower right')

    # Ajustar el diseño y mostrar el gráfico
    plt.tight_layout()
    plt.show()
````
