```python
import spacy
from collections import Counter
import string
```

```python
# Modelo de idioma en español
nlp = spacy.load("es_core_news_sm")
```

```python
# 1. Tokeniza el texto y cuenta palabras (sin símbolos de puntuación)
with open('ramoAzul.txt', 'r', encoding='utf-8') as file:
    text = file.read()
doc = nlp(text)
words = [token.text.lower() for token in doc if token.is_alpha]
num_words = len(words)

print("Número de palabras (sin símbolos de puntuación):", num_words)
```

```python
# 2. Para cada palabra, da el lema y la raíz
lemmas_and_roots = [(token.lemma_, token.lemma_) for token in doc if token.is_alpha]

print("Lema y raíz de cada palabra:", lemmas_and_roots)
```

```python
# 3. ¿Cuántas palabras diferentes hay?
unique_words = set(words)
num_unique_words = len(unique_words)

print("Número de palabras diferentes:", num_unique_words)
```

```python
# 4. Después de lematizar las palabras, ¿cuántas palabras diferentes hay?
unique_lemmas = set(lemma for lemma, _ in lemmas_and_roots)
num_unique_lemmas = len(unique_lemmas)

print("Número de lemas diferentes:", num_unique_lemmas)
```

```python
# 5. ¿Cuál es la diversidad léxica del texto dado?
lexical_diversity = num_unique_words / num_words

print("Diversidad léxica:", lexical_diversity)
```

```python
# 6. ¿Cuáles son las 20 palabras léxicas más frecuentes en el texto? ¿Cuál es su frecuencia?
word_freq = Counter(words)
top_20_words = word_freq.most_common(20)

print("20 palabras léxicas más frecuentes:", top_20_words)
```

```python
# 7. ¿Cuál es el número promedio de palabras por oración?
num_sentences = len(list(doc.sents))
avg_words_per_sentence = num_words / num_sentences

print("Promedio de palabras por oración:", avg_words_per_sentence)
```

```python
# 8. ¿Cuál es la frecuencia de sustantivos, adjetivos y verbos en el texto?
noun_freq = len([token for token in doc if token.pos_ == 'NOUN'])
adj_freq = len([token for token in doc if token.pos_ == 'ADJ'])
verb_freq = len([token for token in doc if token.pos_ == 'VERB'])

print("Frecuencia de sustantivos, adjetivos y verbos:", noun_freq, adj_freq, verb_freq)
```

```python
# Diccionario de mapeo de etiquetas POS
pos_mapping = {
    "determinante": "DET",
    "sustantivo": "NOUN",
    "adjetivo": "ADJ",
    "puntuación": "PUNCT",
    "verbo": "VERB",
    "nombre propio": "PROPN",
    "preposición": "ADP"
    # ...
}
```

```python
with open('GoldStandard.txt', 'r', encoding='utf-8') as file:
    reference_lines = file.readlines()

# Procesa el archivo de referencia y crea una lista de etiquetas POS
reference_pos = [line.strip().split()[-1] for line in reference_lines]

# Procesa el texto con spaCy y crea una lista de etiquetas POS
doc = nlp(text)
spacy_pos = [token.pos_ for token in doc]

# Compara las etiquetas POS de spaCy con las del archivo de referencia
for i in range(len(reference_pos)):
    reference_tag = reference_pos[i]
    spacy_tag = spacy_pos[i]
    if spacy_tag == pos_mapping.get(reference_tag, "NO_MAPEADO"):
        print(f"Token: {doc[i].text}, POS spaCy: {spacy_tag}, POS Referencia: {reference_tag}, Coincide: Sí")
    else:
        print(f"Token: {doc[i].text}, POS spaCy: {spacy_tag}, POS Referencia: {reference_tag}, Coincide: No")
```

```python
# Cuenta cuántas etiquetas coinciden y cuántas no coinciden
coincidencias = 0
no_coincidencias = 0

for i in range(len(reference_pos)):
    reference_tag = reference_pos[i]
    spacy_tag = spacy_pos[i]
    if spacy_tag == pos_mapping.get(reference_tag, "NO_MAPEADO"):
        coincidencias += 1
    else:
        no_coincidencias += 1

# Calcula el accuracy
accuracy = coincidencias / (coincidencias + no_coincidencias)

print(f"Etiquetas coincidentes: {coincidencias}")
print(f"Etiquetas no coincidentes: {no_coincidencias}")
print(f"Accuracy: {accuracy * 100:.2f}%")
```
