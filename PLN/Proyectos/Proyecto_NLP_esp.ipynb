#### Dice Score

````python
import re

def preprocess_text(text):
    # Eliminar puntuación y convertir a minúsculas
    text = re.sub(r'[^\w\s]', '', text.lower())
    return text

def dice_score(s1, s2):
    s1 = preprocess_text(s1)
    s2 = preprocess_text(s2)

    s1_split = set(re.findall(r"\w+|[^\w\s]", s1, re.UNICODE))
    s2_split = set(re.findall(r"\w+|[^\w\s]", s2, re.UNICODE))

    dice_coef = len(s1_split.intersection(s2_split)) / (len(s1_split) + len(s2_split))
    return round(dice_coef, 2)

# Aplicar la función a cada par de frases en tu DataFrame
dice_scores = [dice_score(row['Frase1'], row['Frase2']) for index, row in frases_espp.iterrows()]

# Crear un dataframe con los coeficientes de Dice
sim_esp_s0d = pd.DataFrame({'Dice Coefficient': dice_scores})

# Adjuntar los coeficientes de Dice al dataframe original
sim_esp_s0d = pd.concat([frases_espp[['Frase1', 'Frase2']], sim_esp_s0d], axis=1)
````

#### Distancia Euclidiana

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import euclidean_distances
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer

# Configurar el stemmer para español
stemmer = SnowballStemmer('spanish')

# Aplicar stemming a las columnas
frases_espp['Frase1_stemmed_ssc'] = frases_espp['Frase1'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))
frases_espp['Frase2_stemmed_ssc'] = frases_espp['Frase2'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))

# Crear el objeto vectorizador TF-IDF. Usaremos unigramas y bigramas.
tfidf = TfidfVectorizer(ngram_range=(1, 2,))

# Ajustar el vectorizador TF-IDF en todas las oraciones y transformar cada conjunto por separado
matriz_f1 = tfidf.fit_transform(frases_espp['Frase1_stemmed_ssc'])
matriz_f2 = tfidf.transform(frases_espp['Frase2_stemmed_ssc'])

# Calcular la distancia euclidiana para cada par de oraciones correspondiente
euclidean_distances = euclidean_distances(matriz_f1, matriz_f2)

# Crear un DataFrame con las distancias euclidianas
dist_esp_sse = pd.DataFrame(euclidean_distances.diagonal(), columns=['Euclidean Distance'])

# Ajustar el parámetro alpha
alpha = 0.001

# Calcular la similitud utilizando la función exponencial inversa
dist_esp_sse['Similitud'] = np.exp(-alpha * dist_esp_sse['Euclidean Distance'])

# Normalizar la similitud a una escala de 0 a 1
dist_esp_sse['Similitud'] = ( (dist_esp_sse['Similitud'] - dist_esp_sse['Similitud'].min()) / (dist_esp_sse['Similitud'].max() - dist_esp_sse['Similitud'].min()) )

# Adjuntar las distancias al DataFrame original para una mejor visualización
dist_esp_sse = pd.concat([frases_espp[['Frase1', 'Frase2']], dist_esp_sse], axis=1)
````

#### Distancia Jaccard

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import pairwise_distances
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer

# Configurar el stemmer para español
stemmer = SnowballStemmer('spanish')

# Aplicar stemming a las columnas
frases_espp['Frase1_stemmed_ssc'] = frases_espp['Frase1'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))
frases_espp['Frase2_stemmed_ssc'] = frases_espp['Frase2'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))

# Crear el objeto vectorizador TF-IDF. Usaremos unigramas y bigramas.
tfidf = TfidfVectorizer(ngram_range=(1, 2,))

# Ajustar el vectorizador TF-IDF en todas las oraciones y transformar cada conjunto por separado
matriz_f1 = tfidf.fit_transform(frases_espp['Frase1_stemmed_ssc'])
matriz_f2 = tfidf.transform(frases_espp['Frase2_stemmed_ssc'])

# Convertir las matrices dispersas a matrices densas
matriz_f1_dense = matriz_f1.toarray()
matriz_f2_dense = matriz_f2.toarray()

# Calcular la distancia de jaccard para cada par de oraciones
jaccard_distances = pairwise_distances(matriz_f1_dense, matriz_f2_dense, metric='jaccard')

# Crear un DataFrame con las distancias
dist_esp_ssj = pd.DataFrame(jaccard_distances.diagonal(), columns=['Jaccard Distance'])

# Crear un DataFrame con las similitudes invertidas de Jaccard
dist_esp_ssj['Jaccard Similarity'] = pd.DataFrame(1 - jaccard_distances.diagonal(), columns=['Jaccard Similarity'])

# Adjuntar las distancias al DataFrame original para una mejor visualización
dist_esp_ssj = pd.concat([frases_espp[['Frase1', 'Frase2']], dist_esp_ssj], axis=1)
````

````python
# Función para calcular la similitud entre frases
def spacy_similarity(s1, s2):
    # Procesar las frases con el modelo de SpaCy
    doc1 = nlps(s1)
    doc2 = nlps(s2)

    # Calcular la similitud entre los documentos
    similarity = doc1.similarity(doc2)

    # Escalar la similitud al rango [0, 1]
    scaled_similarity = (similarity + 1) / 2

    return round(scaled_similarity, 2)

# Calcular la similitud para cada par de frases
similarity_scores_spacy = [spacy_similarity(frases_espp['Frase1'].iloc[i], frases_espp['Frase2'].iloc[i]) for i in range(len(frases_espp))]

# Crear un DataFrame con las puntuaciones de similitud
spacy_sim = pd.DataFrame({'SpaCy Similarity': similarity_scores_spacy})

# Adjuntar las puntuaciones de similitud al DataFrame original
spacy_sim = pd.concat([frases_espp[['Frase1', 'Frase2']], spacy_sim], axis=1)
````

````python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Cargar el modelo BERT preentrenado en español
model_name = 'paraphrase-xlm-r-multilingual-v1'
model = SentenceTransformer(model_name)

# Calcular la similitud entre frases
def bert_similarity(s1, s2):
    # Obtener las representaciones vectoriales de las frases
    embeddings = model.encode([s1, s2], convert_to_tensor=True)

    # Calcular la similitud coseno entre las representaciones vectoriales de las frases
    similarity = util.pytorch_cos_sim(embeddings[0], embeddings[1]).item()

    # Ajustar la escala al rango [0, 1]
    scaled_similarity = (similarity + 1) / 2

    return round(scaled_similarity, 2)

# Calcular la similitud para cada par de frases
similarity_scores = [bert_similarity(frases_espp['Frase1'].iloc[i], frases_espp['Frase2'].iloc[i]) for i in range(len(frases_espp))]

# Crear un DataFrame con las puntuaciones de similitud
bert_00c = pd.DataFrame({'BERT Similarity': similarity_scores})

# Adjuntar las puntuaciones de similitud al DataFrame original
bert_00c = pd.concat([frases_espp[['Frase1', 'Frase2']], bert_00c], axis=1)
````

### Similitud coseno

````python
# Cargar el modelo entrenado
model = Word2Vec.load("word2vec_model.model")

def word2vec_similarity(s1, s2):
    s1_tokens = word_tokenize(s1.lower())
    s2_tokens = word_tokenize(s2.lower())

    # Filtrar palabras que no están en el vocabulario del modelo
    s1_tokens = [word for word in s1_tokens if word in model.wv.key_to_index]
    s2_tokens = [word for word in s2_tokens if word in model.wv.key_to_index]

    # Calcular la similitud coseno entre los vectores promedio de las palabras
    if s1_tokens and s2_tokens:
        similarity = model.wv.n_similarity(s1_tokens, s2_tokens)
        # Mapear al rango [0, 1]
        scaled_similarity = (similarity + 1) / 2
        return round(scaled_similarity, 2)
    else:
        return 0.0  # Manejar el caso donde ninguna palabra está en el vocabulario

# Calcular la similitud para cada par de frases
similarity_scores = [word2vec_similarity(frases_espp['Frase1'].iloc[i], frases_espp['Frase2'].iloc[i]) for i in range(len(frases_espp))]

# Crear un DataFrame con las puntuaciones de similitud
word2vec_p0c = pd.DataFrame({'Word2Vec Similarity': similarity_scores})

# Adjuntar las puntuaciones de similitud al DataFrame
word2vec_p0c = pd.concat([frases_espp[['Frase1', 'Frase2']], word2vec_p0c], axis=1)
````

## TF-IDF

Como funciona TfidfVectorizer

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# Ejemplo de documentos de texto
documents = ["Este es el primer documento.",
             "Este documento es el segundo documento.",
             "Y este es el tercer documento.",
             "¿Es este el primer documento?"]

# Crear un objeto TfidfVectorizer con tokenización y bigramas
vectorizer = TfidfVectorizer(ngram_range=(1, 2))

# Ajustar y transformar los documentos
tfidf_matrix = vectorizer.fit_transform(documents)

# Obtener el vocabulario (términos) y la matriz TF-IDF
feature_names = vectorizer.get_feature_names_out()
print("Vocabulario (términos):", feature_names)
print("Matriz TF-IDF:\n", tfidf_matrix.toarray())
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# Ejemplo de documentos de texto que incluyen signos de puntuación
documents = ["Este es un ejemplo de texto con signos de puntuación.",
             "¡Los signos de puntuación también son importantes!"]

# Usar TfidfVectorizer con token_pattern personalizado que incluye signos de puntuación
vectorizer = TfidfVectorizer(token_pattern=r'\b\w+\b|[\.,;\'"¡!\?]',  # Incluye signos de puntuación como tokens
                             stop_words='english')  # Opcional: quitar stop words en inglés

# Ajustar y transformar los documentos
tfidf_matrix = vectorizer.fit_transform(documents)

# Obtener el vocabulario (términos) y la matriz TF-IDF
feature_names = vectorizer.get_feature_names_out()
print("Vocabulario (términos):", feature_names)
print("Matriz TF-IDF:\n", tfidf_matrix.toarray())
````

## Stemming con NLTK

### Ingles

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Ejemplo de documentos de texto
documents = ["running runner ran", "cook, cooking, cooker, cookies, cake"]

# Stemming con NLTK
stemmer = PorterStemmer()
stemmed_documents = [' '.join([stemmer.stem(word) for word in word_tokenize(doc)]) for doc in documents]

# Usar TfidfVectorizer después del stemming
vectorizer = TfidfVectorizer(token_pattern=r'\b\w+\b|[\.,;\'"¡!\?]')
tfidf_matrix = vectorizer.fit_transform(stemmed_documents)

# Obtener el vocabulario (términos) y la matriz TF-IDF
feature_names = vectorizer.get_feature_names_out()
print("Vocabulario (términos):", feature_names)
print("Matriz TF-IDF:\n", tfidf_matrix.toarray())
````

#### Con puntuación

````python
import pandas as pd
import nltk
from nltk.stem import SnowballStemmer
from nltk.tokenize import RegexpTokenizer

# Descargar el recurso necesario para NLTK
nltk.download('punkt')

# Crear un DataFrame de ejemplo
data = {'Texto': ["corriendo, corredor corrió!", "jugando jugó jugará."]}
df = pd.DataFrame(data)

# Configurar el stemmer para español
stemmer = SnowballStemmer('spanish')

# Tokenización con una expresión regular que incluye signos de puntuación
tokenizer = RegexpTokenizer(r'\w+|[^\w\s]')
df['Texto_stemmed'] = df['Texto'].apply(lambda x: ' '.join([stemmer.stem(word) for word in tokenizer.tokenize(x)]))

# Mostrar el DataFrame resultante
print(df)
````

## Lematización con spaCy

### Inglés

````python
from sklearn.feature_extraction.text import TfidfVectorizer
import spacy

# Ejemplo de documentos de texto
documents = ["running runner ran", "cook, cooking, cooker, cookies, cake"]

# Lematización con spaCy
nlp = spacy.load("en_core_web_sm")
lemmatized_documents = [' '.join([token.lemma_ for token in nlp(doc)]) for doc in documents]

# Usar TfidfVectorizer después de la lematización
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(lemmatized_documents)

# Obtener el vocabulario (términos) y la matriz TF-IDF
feature_names = vectorizer.get_feature_names_out()
print("Vocabulario (términos):", feature_names)
print("Matriz TF-IDF:\n", tfidf_matrix.toarray())
````
